```python
import requests
from yaml import load, FullLoader
```

```python
API_URL = 'https://api-metrika.yandex.ru/stat/v1/data'
```

```python
# период отчета
start_date = '2023-10-01'
end_date = '2023-10-07'

# 44147844 - демо-счетчик Метрики
counter = 44147844
```

```python
with open('test_config.yaml') as f:
    config = load(f, Loader=FullLoader)
    
token = config['metrika']['token']  # коллега сам подставляет свой токен в yaml-файл
```

```python
# для доступа к демо-счетчику нужен пустой заголовок веб-запроса
headers = ''

# для доступа к вашему закрытому счетчику используейте заголовок
# headers = {'Authorization': 'OAuth {}'.format(token)}
# документация: https://yandex.ru/dev/metrika/doc/api2/intro/authorization-docpage/
```

```python
# задаем параметры запроса
params = {
            'id': counter,
            'date1': start_date,
            'date2': end_date,
            'metrics': 'ym:s:users',
            'dimensions': 'ym:s:date,ym:s:regionCity',
        }
```

```python
r = requests.get(API_URL, params=params, headers=headers)
r
```

```python
r.status_code
```

```python
r.json()
```

```python
with open('metrika_data.csv', 'w') as f:
    for rec in r.json()['data']:
        f.write('{},{}\n'.format(rec['dimensions'][0]['name'], rec['metrics'][0]))
```
