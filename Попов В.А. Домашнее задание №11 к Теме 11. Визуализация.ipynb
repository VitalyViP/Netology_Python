```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('tmdb_5000_movies.csv')
```

```python
# 10     .   .       .
top_10=df.nlargest(n=10,columns=['budget'])
top_10.plot(kind='bar', x='original_title', y='budget')
plt.title('Budget for Movies (TOP 10)')
plt.xlabel('Movie Title')
plt.ylabel('Budget')
plt.show()
```

```python
#      .   .      .
df['vote_average'].plot(kind='hist', bins=10)
plt.title('Votes Distribution')
plt.xlabel('Votes')
plt.ylabel('Frequency')
plt.show()
```

```python
#  10         .   .           .
top_10=df.nlargest(n=10,columns=['revenue'])
top_10.plot(kind='scatter', x='budget', y='revenue')
plt.title('Revenue vs. Budget')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
#     .   .  
df['vote_average'].plot(kind='density')
plt.title('Density Plot of Vote Average')
plt.xlabel('vote_average')
plt.ylabel('Density')
plt.show()
```

```python
#          .   .        . 
data = df['production_countries'].apply(lambda x: eval(x))

us_count = sum(1 for entry in data if any(country.get("iso_3166_1") == "US" for country in entry))
non_us_count = len(data) - us_count

labels = 'US', 'Non-US'
sizes = [us_count, non_us_count]
colors = ['green', 'grey']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Proportion of US vs Non-US production countries')

plt.show()
```
