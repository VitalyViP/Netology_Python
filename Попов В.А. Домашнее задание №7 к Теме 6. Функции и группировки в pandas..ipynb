```python
# Задание1
import pandas as pd

def classified(rating):
    if rating <= 2:
        return 'low'
    elif rating <= 4:
        return 'middle'
    else:
        return 'high'


df_movies=pd.read_csv('movies.csv', usecols=['movieId','title'])
df_ratings=pd.read_csv("ratings.csv", usecols=['movieId','rating'])
df_ratings = df_ratings.groupby('movieId').mean().reset_index()
df_class=df_ratings.merge(df_movies, on='movieId', how='left')
df_class['class'] = df_class['rating'].apply(classified)
df_class.head()
```

```python
# Задание2
import pandas as pd
df_location = pd.read_csv('keywords.csv')
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
def sorting(row):
    keyword_set = set(row['keyword'].lower().split())
    for region, city_list in geo_data.items():
        if keyword_set.intersection(city_list):
            return region
    return 'undefined'

df_location['location'] = df_location.apply(sorting, axis=1)
df_location[df_location['location'] !='undefined'].head()
```
