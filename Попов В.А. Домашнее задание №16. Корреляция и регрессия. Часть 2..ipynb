```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('water.csv')
df.head()
```

```python
# Задача 1
x = df[['hardness']]
y = df['mortality']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)  # обучается на данных и подбирает оптимальные коэффициенты
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
# Значение коэффициента детерминации
model.score(x_test, y_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, c='r')
```

```python
# График остатков 1
plt.scatter(y_pred, (y_test - y_pred))
plt.plot(y_pred, (y_test - y_pred), c='r')
```

```python
# График остатков 2
plt.hist(y_test - y_pred, bins=20)
```

```python
# Ответ задание 1. Из визуализаций остатков видна хаотичность их значений, 
# что может указывать на слабую прогнозную составляющую метода линейной регрессии, 
# Значение r2 ниже 0.5 означает, что модель плохо аппроксимирует данные и не может сделать точные прогнозы.
# Такие модели не рекомендуется использовать в прогностических задачах или принимать на основе их результатов решения.
```
