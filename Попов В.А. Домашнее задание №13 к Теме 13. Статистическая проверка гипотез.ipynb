# Задачи

<div class="alert alert-info">
<b>Задание №1</b>
    
Профессор Оук подозревает, что у покемонов в классе `grass` более сильная обычная атака, чем у покемонов в классе `rock`. Проверьте, прав ли он, и статистически убедите его в своём выводе.
    
    
Примечание: если есть покемоны, которые относятся к обоим классам, отбросьте их.
    
Вы можете предположить, что распределение обычных атак является нормальным для всех классов покемонов.

</div>


```python
filtered_pokemon = pokemon[pokemon['Class 1'].isin(['Grass', 'Rock']) | pokemon['Class 2'].isin(['Grass', 'Rock'])].drop_duplicates()

grass_pokemon = filtered_pokemon[(filtered_pokemon['Class 1'] == 'Grass') | (filtered_pokemon['Class 2'] == 'Grass')]
rock_pokemon = filtered_pokemon[(filtered_pokemon['Class 1'] == 'Rock') | (filtered_pokemon['Class 2'] == 'Rock')]

t_stat, p_value = ttest_ind(grass_pokemon['Attack'], rock_pokemon['Attack'], equal_var=False)

if p_value < 0.05:
    print("Подозрение профессора Оука статистически значимо. Покемоны класса Grass имеют значительно более сильную обычную атаку, чем покемоны класса Rock")
else:
    print("Существенной разницы в обычной атаке между покемонами классов Grass и Rock нет.")
```

<div class="alert alert-info">
<b>Задание №2</b>
    
Профессор Оук уже долго не может спать по ночам, ведь его волнует вопрос, правда ли, что покемоны в классе `Water` в среднем быстрее, чем покемоны в классе `Normal`.
    
    
Проверьте, прав ли он, и убедите его в своём выводе статистически.
    
Примечание: если есть покемоны, которые относятся к обоим классам, отбросьте их.
    
Вы можете предположить, что распределение скорости движения является нормальным для всех классов покемонов.

</div>

```python
filtered_pokemon = pokemon[pokemon['Class 1'].isin(['Water', 'Normal']) | pokemon['Class 2'].isin(['Water', 'Normal'])].drop_duplicates()

water_pokemon = filtered_pokemon[(filtered_pokemon['Class 1'] == 'Water') | (filtered_pokemon['Class 2'] == 'Water')]
normal_pokemon = filtered_pokemon[(filtered_pokemon['Class 1'] == 'Normal') | (filtered_pokemon['Class 2'] == 'Normal')]

t_stat, p_value = ttest_ind(water_pokemon['Speed'], normal_pokemon['Speed'], equal_var=False)

if p_value < 0.05:
    print("Профессор Оук, на основании статистического анализа, выяснилось, что покемоны класса Water в среднем быстрее покемонов класса Normal.")
else:
    print("Профессор Оук, статистический анализ не показывает существенной разницы в скорости между покемонами классов Water и Normal.")
```

<div class="alert alert-info">
<b>Задание №3</b>
    
Профессор Оук тот ещё безумец. Он изобрел сыворотку, способную ускорить покемона. Но мы усомнились в эффективности его вакцины. Профессор дал эту сыворотку следующим покемонам: см. массив `treathed_pokemon`. Проверьте, работает ли его сыворотка вообще, и убедите всех в своём выводе статистически.
    
    
Вы можете предположить, что распределение скорости движения является нормальным для всех классов покемонов.

</div>

```python
treathed_pokemon = ['Mega Beedrill', 'Mega Alakazam', 
                    'Deoxys Normal Forme', 'Mega Lopunny']

avg_treated_speed = pokemon[pokemon['Name'].isin(treathed_pokemon)]['Speed'].mean()

population_speed = pokemon['Speed']
t_stat, p_value = ttest_1samp(population_speed, avg_treated_speed)

if p_value < 0.05:
    print("Эффективность сыворотки статистически значима. Обработанные покемоны демонстрируют значительное увеличение скорости по сравнению с общей популяцией.")
else:
    print("По статистике, сыворотка не оказывает существенного влияния на скорость обработанных покемонов.")
```

<div class="alert alert-info">
<b>Задание №4</b>
    
Профессор Оук всегда любил истории о легендарных покемонах. Но профессор не очень уверен, что они лучшие из всех покемонов. Оук предложил нам в этом разобраться. Проверьте, действительно ли сумма характеристик `HP`,`Attack`,`Defense` у легендарных покемонов выше, чем у других покемонов. А произведение этих же параметров? Найдите ответы на эти вопросы и убедите всех в своём выводе статистически.
   

Вы можете предположить, что распределение сумм и произведений этих параметров является нормальным для всех классов покемонов.

</div>

```python
df = pd.DataFrame(pokemon)

legendary_pokemon = df[df['Legendary'] == True].copy()
non_legendary_pokemon = df[df['Legendary'] == False].copy()

legendary_pokemon.loc[:, 'Sum'] = legendary_pokemon['HP'] + legendary_pokemon['Attack'] + legendary_pokemon['Defense']
legendary_pokemon.loc[:, 'Product'] = legendary_pokemon['HP'] * legendary_pokemon['Attack'] * legendary_pokemon['Defense']

non_legendary_pokemon.loc[:, 'Sum'] = non_legendary_pokemon['HP'] + non_legendary_pokemon['Attack'] + non_legendary_pokemon['Defense']
non_legendary_pokemon.loc[:, 'Product'] = non_legendary_pokemon['HP'] * non_legendary_pokemon['Attack'] * non_legendary_pokemon['Defense']

t_statistic_sum, p_value_sum = stats.ttest_ind(legendary_pokemon['Sum'], non_legendary_pokemon['Sum'])
t_statistic_product, p_value_product = stats.ttest_ind(legendary_pokemon['Product'], non_legendary_pokemon['Product'])

print("Результаты теста для суммы характеристик (HP, Attack, Defense):")
print("t-статистика:", t_statistic_sum)
print("p-значение:", p_value_sum)
print("Результаты теста для произведения характеристик (HP, Attack, Defense):")
print("t-статистика:", t_statistic_product)
print("p-значение:", p_value_product)
```

```python
# Исходя из предоставленных результатов теста, можно сделать следующие выводы:
# Для суммы характеристик (HP, Attack, Defense):
# Значение t-статистики равно 11.59, что указывает на статистическую значимость различий между средними суммами характеристик у легендарных и обычных покемонов. 
# p-значение составляет 7.97e-29, что крайне мало, указывая на то, что такое экстремальное наблюдаемое значение суммы характеристик было бы очень маловероятно при условии верности нулевой гипотезы.

# Для произведения характеристик (HP, Attack, Defense):
# Значение t-статистики равно 13.26, также указывая на статистическую значимость различий между средними произведениями характеристик у легендарных и обычных покемонов. 
# p-значение равно 1.99e-36, подтверждая, что такое экстремальное наблюдаемое значение произведения характеристик было бы очень маловероятно при условии верности нулевой гипотезы.
# Таким образом, на основании статистических данных можно сделать вывод, что и сумма, и произведение характеристик HP, Attack, Defense у легендарных покемонов значительно выше, чем у других покемонов, и эти различия являются статистически значимыми.
```

Профессор Оук частенько наблюдает за боями покемонов. После очередных таких боёв Оук выделил три класса `best_defence_class`, которые, на его взгляд, одинаковы по силе обычной защиты `Defense`. Проверьте, действительно ли эти классы покемонов не отличаются по уровню защиты статистически значимо. Всё та же статистика вам в помощь!
   

Вы можете предположить, что распределение сумм и произведений этих параметров является нормальным для всех классов покемонов.

</div>

```python
# best_defence_class = ['Rock', 'Ground', 'Steel', 'Ice']
# best_defence_class
```

```python
df = pd.DataFrame(pokemon)

best_defence_class = ['Rock', 'Ground', 'Steel', 'Ice']
defense_data = [df[(df['Class 1'] == cls) | (df['Class 2'] == cls)]['Defense'] for cls in best_defence_class]

# Performing ANOVA test
f_statistic, p_value = f_oneway(*defense_data)

# Checking the p-value to determine statistical significance
if p_value < 0.05:
    print("Профессор Оук, на основании статистического анализа, выяснилось, что характеристики защиты указанных классов статистически различаются.")
else:
    print("Профессор Оук, статистический анализ показывает, что разницы в характеристиках защиты указанных классов нет.")
```
