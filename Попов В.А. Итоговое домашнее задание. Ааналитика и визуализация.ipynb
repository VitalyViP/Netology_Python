```python
import numpy as np
import scipy.stats as st 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Netology_Data')
array_18 = df.iloc[:,18]
array_18.head(5)
```

```python
#Статистики

mean = np.mean(array_18) #Среднее значение выборки
D = np.var(array_18) #Дисперсия выборки
std = np.std(array_18) #Стандартное отклонение
mean,D,std
```

```python
#Доверительный интервал

alpha =0.05
right = np.mean(array_18)+st.t.ppf(1-alpha/2,len(array_18)-1)*np.std(array_18,ddof=1)/np.sqrt(len(array_18))
left = np.mean(array_18)-st.t.ppf(1-alpha/2,len(array_18)-1)*np.std(array_18,ddof=1)/np.sqrt(len(array_18))
left,right
```

```python
#Рисуем график

plt.figure(figsize = (10,7))
sns.histplot(array_18)
plt.axvline(x=mean, color='red', linestyle='--', label='Среднее значение')
plt.axvline(x=left, color='green', linestyle='--', label='95% Доверительный интервал (нижний предел)')
plt.axvline(x=right, color='green', linestyle='--', label='95% Доверительный интервал (верхний предел)')

plt.legend(loc='upper right')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Распределение')
    
plt.show()
```
